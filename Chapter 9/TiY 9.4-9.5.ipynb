```python
class Restaurant():
    """Building my restaurant."""
    
    def __init__(self,name,cuisine,):
        """Initianlize restaurant name and type of food."""
        
        self.name = name
        self.cuisine = cuisine
        self.number_served = 0
        
    def describe_restaurant(self):
        """Print message describing the restaurant."""
        
        describe = self.name.title() + " is a restaurant that serves " + self.cuisine + " food."
        
        return describe
    
    def open_restaurant(self):
        """Print message opening restarant"""
        
        print(self.name.title() + " is open for business.")
        
    def read_served(self):
        """Print the number served until this point."""
        print('We have served ' + str(self.number_served) + 
              " until this point.")
        
    def set_number_served(self,served):
        """The number of served customers."""
        self.number_served = served
        
    def increment_number_served(self,service):
        """Increase the number served."""
        self.number_served += service
        
restaurant = Restaurant('Pappas','Seafood')

```

```python
print(restaurant.describe_restaurant())
restaurant.number_served = 15
print('\n')
restaurant.read_served()

restaurant.set_number_served(48)
restaurant.read_served()
restaurant.increment_number_served(10)
restaurant.read_served()
```

```python
class User():
    """Create user attributes"""
    def __init__(self,first_name,last_name,height,sex,race):
        
        """Initialize user attributes"""
        self.first_name = first_name
        self.last_name = last_name        
        self.height = height
        self.sex = sex
        self.race = race
        self.login_attempts = 0
        
    def describe_user(self):
        """Print summary of user"""
        print("\nHere's a brief summary of " + self.first_name.title() + ":")
        print("Name: " + self.first_name + " " + self.last_name)
        print("Height: " + self.height)
        print("Sex: " + self.sex)
        print("Race: " + self.race)
        
    def greet_user(self):
        """Print a greeting to user."""
        print('\n' + self.first_name.title() + ' ' +  self.last_name.title() + 
              ' welcome to the village.')
        
    def increment_login_attempts(self):
        """Increment the login attempts by 1"""
        self.login_attempts += 1
        
    def reset_login_attempts(self):
        """reset the login attempts to zero."""
        self.login_attempts = 0
        
        
bob = User('Bobby','Knight','6foot','male','white')

bob.describe_user()
bob.greet_user()

tim = User('Time','Farris','7foot','male','black')

tim.describe_user()
tim.greet_user()

jg = User('J','G','6foot5inch','male','black')
jg.describe_user()
jg.greet_user()
jg.increment_login_attempts()
jg.increment_login_attempts()
jg.increment_login_attempts()
jg.increment_login_attempts()
print('You have tried to login ' + str(jg.login_attempts) + ' times.')
jg.reset_login_attempts()
print('\nYour logins have been reset')
print('You have tried to login ' + str(jg.login_attempts) + ' times.')
```
