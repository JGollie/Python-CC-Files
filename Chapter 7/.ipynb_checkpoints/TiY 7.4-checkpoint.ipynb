```python
prompt = "\nEnter the toppings you would like on your pizza:"
prompt += "\nEnter 'done' when finished adding toppings to your pizza:"

active = True
while active:
    topping = input(prompt)
    
    if topping == 'done':
        print("Your pizza is in que to be cooked in the oven.")
        active = False
    else: 
        print(topping)
```

```python
prompt = "\nWelcome to JLG Theaters. How old are you?"
prompt += "\nTicket prices are:"
prompt += "\n\tChild under 3: Free \n\t Ages 3 - 12: $10 \n\t Over 12: $15"
prompt += "\n"

age = int()

while int(age) != 484848:    
    age = input(prompt)
    if int(age) <= 3:
        print("\nYour ticket is Free.")
    elif int(age) <= 12:
        print("\nYour ticket will be $10")
    elif int(age) == 484848:
        break
    else:
        print("\nYour ticket will be $15")
```
